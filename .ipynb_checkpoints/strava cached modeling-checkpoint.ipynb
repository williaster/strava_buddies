```python
# Look at segments per mile 
def get_segments_per_mile(conn, athlete_id, n_activities, act_type=""):
    activities = buds.get_user_activity_options(conn, athlete_id, n_activities, act_type=act_type)
    metrics = { act["id"] : { "distance" : act["distance"], 
                              "seg_ids"  : len(buds.get_unique_segments_for_activities(conn, 
                                                                                  athlete_id, 
                                                                                  [act["id"]]))} \
                for act in activities }
    ratios = { act_id : metrics[act_id]["seg_ids"] / float(metrics[act_id]["distance"]) 
               for act_id in metrics.keys() }
    
    return pd.DataFrame(ratios.values(), index=ratios.keys())
```

```python
df_n_seg_per_mile_ride = get_segments_per_mile(conn, nicole, 200, act_type="Ride")
df_n_seg_per_mile_run  = get_segments_per_mile(conn, nicole, 200, act_type="Run")
```

```python
# Different distance metrics
def euclid(other, athlete, friend_col):
    return pd.Series([other.ix[friend_col],
                      distance.euclidean(other.ix["ride_count":"annual_dist_std"], 
                                         athlete.ix["ride_count":"annual_dist_std"])],
                      index=["friend", "distance"])

# euclidean is additive anyway, but exploring using week as a single unit
# this also supports weights
def w_euclid(other, athlete, friend_col, weights):
    delta = other.ix["ride_count":"annual_dist_std"] - \
            athlete.ix["ride_count":"annual_dist_std"]

    return pd.Series([other.ix[friend_col], np.sqrt((weights*delta*delta).sum())],
                      index=["friend", "distance"])

def cosine(other, athlete, friend_col):
    return pd.Series([other.ix[friend_col],
                      distance.cosine(other.ix["ride_count":"annual_dist_std"], 
                                      athlete.ix["ride_count":"annual_dist_std"])],
                      index=["friend", "distance"])

def w_cosine(other, athlete, friend_col, weights):
    dist_ride = distance.cosine(other.ix["ride_count"], athlete.ix["ride_count"])
    dist_run  = distance.cosine(other.ix["run_count"], athlete.ix["run_count"])
    dist_dow  = distance.cosine(other.ix["mon_freq":"sun_freq"], 
                                athlete.ix["mon_freq":"sun_freq"])
    dist_dist = distance.cosine(other.ix["annual_dist_median"], athlete.ix["annual_dist_median"])
    dist_var  = distance.cosine(other.ix["annual_dist_std"], athlete.ix["annual_dist_std"])
    
    vals = np.array([dist_ride, dist_run, dist_dow, dist_dist, dist_var])
    return pd.Series([other.ix[friend_col], np.sum( weights * vals )],
                      index=["friend", "distance"])
```
