```python
soup = bs4.BeautifulSoup(req_athlete.text)
```

```python
row      = soup.select(".activity-calendar .row")[0]
all_days = row.select(".day")
act_days = set(row.select(".highlighted"))

def get_freq_by_day(soup):
    rows = soup.select(".activity-calendar .row")[1:-1] # 0  = headers
                                                        # -1 = variable
    result = np.ones(shape=(3,7)) # no div by zero, 'pseudo ct'

    for i in range( len(rows) ): # for each week
        all_days      = rows[i].select(".day")
        
        #highlighted = activity
        activity_days = set( rows[i].select(".highlighted") )
        
        # boolean list for where activities occurr
        result[i] = [ day in activity_days for day in all_days ]
        
    return result.sum(axis=0) / result.sum()

get_freq_by_day(soup)
```

```python
def get_run_ride_ct(soup):
    run_icon  = soup.select(".activity-breakdown .icon-run")[0]
    ride_icon = soup.select(".activity-breakdown .icon-ride")[0]

    run_ct  = int( run_icon.previous_sibling.previous_sibling.text )
    ride_ct = int( ride_icon.previous_sibling.previous_sibling.text )
    
    return run_ct, ride_ct

get_run_ride_ct(soup)
```
