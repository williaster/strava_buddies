```python
row      = soup.select(".activity-calendar .row")[0]
all_days = row.select(".day")
act_days = set(row.select(".highlighted"))

def get_freq_by_day(soup):
    rows = soup.select(".activity-calendar .row")[1:-1] # 0  = headers
                                                        # -1 = variable
    result = np.ones(shape=(3,7)) # no div by zero, 'pseudo ct'

    for i in range( len(rows) ): # for each week
        all_days      = rows[i].select(".day")
        
        #highlighted = activity
        activity_days = set( rows[i].select(".highlighted") )
        
        # boolean list for where activities occurr
        result[i] = [ day in activity_days for day in all_days ]
        
    return result.sum(axis=0) / result.sum()

get_freq_by_day(soup)
```

```python
def get_run_ride_ct(soup):
    run_icon  = soup.select(".activity-breakdown .icon-run")[0]
    ride_icon = soup.select(".activity-breakdown .icon-ride")[0]

    run_ct  = int( run_icon.previous_sibling.previous_sibling.text )
    ride_ct = int( ride_icon.previous_sibling.previous_sibling.text )
    
    return run_ct, ride_ct

get_run_ride_ct(soup)
```

```python
def get_mi_per_px(soup):
    get_px    = lambda tick: int(re.findall("[a-z:]*([\d]+)px", tick.get("style"))[0])
    yaxis     = soup.select("ul.y-axis")[0]
    two_ticks = yaxis.findChildren()[:2]
    adj_units = MI_PER_KM if "km" in two_ticks[0].text else 1

    delta_mi  = int( two_ticks[1].text ) * adj_units # tick 1 = 0
    delta_px  = ( get_px(two_ticks[1]) - get_px(two_ticks[0]) )
    return delta_mi / delta_px

get_mi_per_px(soup)
```

```python
mi_per_px = get_mi_per_px(soup)
weekly_ivals = soup.select("li.interval")[:-1] # last is filler
ival = weekly_ivals[0]

def get_annual_distances(ivals, mi_per_px):
    """Returns a numpy array of weekly intervals in mi, for the year
       as well as the first non-zero index for subsequent computations
    """
    dists        = np.array([ get_ival_distance(ival, mi_per_px) \
                              for ival in ivals ])
    idx_non_zero = numpy.argmax( dists > 0 ) #
    return dists, idx_non_zero

def get_ival_distance(ival, mi_per_px):
    """Returns the distance in mi for a single distance interval
    """
    try: 
        ival_bar = ival.select("div.fill")[0]
    except: # no bar
        return 0
    
    get_px   = lambda bar: int(re.findall("[a-z:]*([\d]+)px", 
                                          bar.get("style"))[0])
    px_dist  = get_px(ival_bar)
    return mi_per_px * px_dist

dists, idx_non_zero = get_annual_distances(weekly_ivals, mi_per_px)
np.median(dists[idx_non_zero:]), np.std(dists[idx_non_zero:])
```

```python
def get_name(soup):
    first, last = soup.select("title")[0].text.split("|")[0].split()
    return first.encode("utf-8"), last.encode("utf-8")
get_name(soup)
```

```python
def get_location(soup):
    """Fetches athlete city and state, encodes as utf8 for non-ascii chars
    """
    city, state = soup.select("p.location")[0].text.split(",")
    return city.encode("utf-8"), state.strip().encode("utf-8")

get_location(soup)
```
